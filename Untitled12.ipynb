# Stratified Sampling on Dataset
In your dataset, most of the median_income values are between 1.5 and 6, but a few go much higher  those are outliers. If you split your data randomly, you might end up with a situation where all the rich areas fall into the test set and poor areas in the training set (or vice versa). That would mess up your model's ability to learn properly and give biased results.

To avoid that, you use Stratified Sampling, where you group the income into categories (like 1 = very low income, 2 = low, 3 = mid, etc.), and then split the data in a way that each group appears in the same proportion in both training and test sets. This way, your model sees a balanced view of the real-world data, and it can learn and predict more accurately.

Now, We will be importing a tool from sklearn that allows you to split data based on strata (groups)  like the income_cat column.

We are creating the "splitter object" with these rules:

n_splits=1  We only want to split the dataset once.

test_size=0.2  We want 20% of data for testing, 80% for training.

random_state=42  So the shuffle is always the same every time we  run the code (reproducible).

`for train_index, test_index in split.split(prices, prices["income_cat"]):`

This is where the magic happens:

We tell the splitter: "Split my dataset prices, but make sure the groups in income_cat stay in the same proportion!"

It returns two lists:

train_index: row numbers for training set

test_index: row numbers for test set

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits =1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(prices, prices["income_cat"]):#Split data while keeping income_cat proportions
  strat_train_set = prices.loc[train_index] # Get actual training rows from the original dataset
  strat_test_set = prices.loc[test_index] # Get actual testing rows from the original dataset
print(strat_test_set['income_cat'].value_counts()/len(strat_test_set))  # Check if test set has same income_cat distribution
```

Now we need to remove thee income_cat attribute added by us to get the data back to its form:

Why? You grouped students by grades (A, B, C, D) just to make sure:

Each classroom (training and test set) had the same mix of smart and weak students.

Once the students are placed in each class, you dont need their grade labels anymore  you let the teacher teach without knowing those group tags.

Same thing here!


Why this is GOOD:

-> income_cat helps you make a fair train-test split (so your model doesnt get biased).

-> Once the split is done, you dont want to cheat by letting the model "see" that artificial category.

-> So you drop it, and train on the real features (like total_rooms, median_income, households, etc.)

```python
for set_ in (strat_train_set, strat_test_set): # Loop through both train and test sets
  set_.drop('income_cat', axis=1, inplace = True)  # Remove the helper column 'income_cat' from each set
prices = strat_train_set.copy()  # Make a fresh copy of the cleaned training set to use for further analysis or modeling
```

# Data Preperation
Now this is the most important step before train a machine learning model for the task of house prediction. Now let's perform all the necessary data transformations:

```python
#  Data preparation

# Drop the target label to create feature set
prices = strat_train_set.drop("median_house_value", axis=1)  # drop the label
prices_labels = strat_train_set["median_house_value"].copy()  # keep the label

#  Handle missing values
median = prices["total_bedrooms"].median()  # calculate the median
prices["total_bedrooms"].fillna(median, inplace=True)  # fill missing values with median

# Drop non-numeric column for numerical processing
prices_num = prices.drop("ocean_proximity", axis=1)

#  Feature engineering: adding new attributes using custom transformer
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np  # make sure this is imported!

# Correct column indices (double check with prices_num.columns if unsure)
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):  # allow optional column
        self.add_bedrooms_per_room = add_bedrooms_per_room

    def fit(self, X, y=None):
        return self  # no fitting needed, just transformation

    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]

        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

As you can see, there are many data transformation steps that need to be performed in the correct order. Fortunately, Scikit-Learn provides the Pipeline class to help you with such sequences of transformations. Here is a small pipeline for numeric attributes:

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
num_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
prices_num_tr = num_pipeline.fit_transform(prices_num)

from sklearn.compose import ColumnTransformer

num_attribs = list(prices_num.columns)  #  get numeric column names
cat_attribs = ["ocean_proximity"]       # categorical column

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(prices)  # make sure 'prices' = full dataset
```
